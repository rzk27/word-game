```python
import pandas as pd

data = pd.read_csv("AoA_dataset.csv")

data.columns = ['Word', 'Mean_AoA']

data = data.dropna(subset=["Word", "Mean_AoA"])
data["Word"] = data["Word"].astype(str).str.lower()
data["Mean_AoA"] = data["Mean_AoA"].astype(float)

data = data[data['Word'].str.len() <= 15].reset_index(drop=True)
```

```python
data.to_pickle("data.pkl")
```
